12. How does the elif statement work
-  When we have multiple conditions to check we usually use elif conditional block . It first checks the first if condition and if it is not true then it moves to elif block and check all the elif block before moving to else block.

```python
#Ex-12
Num1 = int(input("Enter 1st Numner "))
Num2 = int(input("Enter 2nd Number "))
Operator = input("Enter the operation You Want to perform ")
if Operator == "+":
    Result = Num1 + Num2
    print(f"{Num1} + {Num2} = {Result}")
elif Operator == "-":
    Result = Num1 - Num2
    print(f"{Num1} - {Num2} = {Result}")
elif Operator == "*":
    Result = Num1 * Num2
    print(f"{Num1} * {Num2} = {Result}")
elif Operator == "/":
    Result = Num1 / Num2
    print(f"{Num1} / {Num2} = {Result}")
elif Operator == "%":
    Result = Num1 % Num2
    print(f"{Num1} % {Num2} = {Result}")
```
